```python
import time
from PIL import Image
import IPython

import geopy
from geopy.geocoders import Nominatim
import folium

def ask_arr():
    arr = str(input('Which arrondissement would you like to eat? \n'))
    lst = ['1','2','3','5','6','8','9','10','11','12','13','14','15','16','18','19','20']
    if arr in  lst:
        return arr
    elif arr is '4' or '7' or '12' or '17': 
        print('Sorry, there is no good restaurant in this area!')
        return ask_arr()

def ask_cat():
    newcat= input('Which kind : design, fast-good, insolite, romantique, tradi, world ? \n').lower()
    lst_cat = ['design', 'fast-good', 'insolite', 'romantique', 'tradi', 'world']
    if newcat in lst_cat:
        return newcat
    elif newcat is not lst_cat:
        print('Sorry can you enter one of the 6.')
        return ask_cat()

#def play_again():
    again = str(input('Do you want to try again ?')).lower()
    if again == 'yes' or again== 'y'or again== 'oui':
        app_resto()
    elif again == 'no' or again=='n'or again=='non' :
       print('Bye bye.. Hope to see you again !')
       return None

def app_resto():
    global arr
    arr=ask_arr()
    global newcat
    newcat=ask_cat()
    filtA=df1[df1.Arr==int(arr)].copy()
    filtAC=filtA.loc[(filtA.Categories.str.contains(str(newcat)))].sort_values(by='price').head(3)
    print('Hmmmm okay')
    try:
        print('Eats possible!!!!')
        print(f'The cheapest {newcat} restaurant in the {filtAC.Arrondissement[0]} arrondissement of Paris is : {filtAC.Nameresto1[0]}\n')
        time.sleep(2)
        print(f'{filtAC.Tagline[0]}\n')
        time.sleep(2)
        print(f'You can eat {filtAC.Timelunch[0]} for {filtAC.price[0]}€ really cheap!\n')
        print(f'The adress is {filtAC.adress[0]}. \n')
        print(f'Call now {filtAC.Telephone[0]} to book a table.')
        IPython.display.Image(filtAC.photos[0], width = 440)
    except:
        print(f'Sorry there is no {newcat} restaurant in the {arr} arrondissement')
        return app_resto()
    

name = input ("What's your name?\n")
print(f'Hello {name}! Welcome on the new app *Eats possible in Paris* \n If you are here means that your are hungry...\n')

app_resto()

filtA=df1[df1.Arr==int(arr)].copy()
filtAC=filtA.loc[(filtA.Categories.str.contains(str(newcat)))].sort_values(by='price').head(3)
display(IPython.display.Image(filtAC.photos[0], width = 440))

geolocator = Nominatim(user_agent="Restos")
location = geolocator.geocode(f'{filtAC.adress[0]} {filtAC.zipcode[0]} Paris', timeout=15)

m = folium.Map(location=[location.latitude, location.longitude],zoom_start=17)
m
#play_again()
```
